##### divide the data into X and y

```python
X=df1.drop(['diagnosis','id'],axis=1)
y=df1['diagnosis']
```

```python
X.head()
```

```python
y.head()
```

##### Splitting data into train and test data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,stratify=y)
```

##### Lets apply random forest model along with cross validation and hyperparameter tutning

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rand_f=RandomForestClassifier()
rand_f
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Number of trees in random forest
# n_estimators = [int(x) for x in np.linspace(start = 100, stop = 200, num = 5)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 50, num = 5)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
random_grid
```

```python
grid=GridSearchCV(rand_f,param_grid=random_grid,cv=10)
grid.fit(X,y)
```

```python
grid.best_params_
```

##### Fitting random forest with best parameters

```python
rand_f=RandomForestClassifier(bootstrap=False,max_depth=40,max_features='sqrt',min_samples_leaf=1
                             ,min_samples_split=2)
rand_f
```

```python
rand_f.fit(x_train,y_train)
```

```python
rand_f.score(x_test,y_test)
```

```python
y_pred=rand_f.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
plot_confusion_matrix(rand_f,x_test,y_test,display_labels=['Malignant','Benign'])
plt.show()
```
